```python
def validate_support_values(file_path=""):
    iqtree_file = "/home/piyumal/PHD/BootstrapSupportPrediction/sample/688_1/688_1.suptree.ete3"
    raxml_file = "/home/piyumal/PHD/BootstrapSupportPrediction/sample/688_1/688_1.newick"

    tree_iqtree = Tree(iqtree_file)

    iqtree_node_traversal = []
    for node in tree_iqtree.traverse():
        edge = node.get_edges()[0]
        clade_1 = list(edge[0])
        clade_2 = list(edge[1])
        node_name = clade_1 + clade_2
        node_name = [node.name for node in node_name]
        node_name.sort()
        node_name = ':'.join(node_name)
        node.name = node_name
        iqtree_node_traversal.append(node.name)

    tree_raxml = Tree(raxml_file)

    raxml_node_traversal = []
    for node in tree_raxml.traverse():
        edge = node.get_edges()[0]
        clade_1 = list(edge[0])
        clade_2 = list(edge[1])
        node_name = clade_1 + clade_2
        node_name = [node.name for node in node_name]
        node_name.sort()
        node_name = ':'.join(node_name)
        node.name = node_name
        raxml_node_traversal.append(node.name)

    iqtree_diff = []
    raxml_diff = []
    for u in raxml_node_traversal:
        if u not in iqtree_node_traversal:
            iqtree_diff.append(u)

    for v in iqtree_node_traversal:
        if v not in raxml_node_traversal:
            raxml_diff.append(v)
    iqtree_diff.reverse()
    raxml_node_traversal_v1 = copy.deepcopy(raxml_node_traversal)
    iqtree_node_traversal_v1 = copy.deepcopy(iqtree_node_traversal)
    for idx_iter, item in enumerate(iqtree_diff):
        index_diff = raxml_node_traversal.index(item)
        raxml_node_traversal[index_diff] = raxml_diff[idx_iter]

    iqtree_idx = []
    raxml_node_traversal_post = raxml_node_traversal
    
    raxml_idx = []
    iqtree_node_traversal_post = iqtree_node_traversal[1:]

    for val in iqtree_node_traversal[1:]:
        iqtree_idx.append(raxml_node_traversal_post.index(val))
    print(raxml_node_traversal)
    print(iqtree_node_traversal)
    print(iqtree_idx)
    print(iqtree_diff)
    print(raxml_diff)
```
